```python
import matplotlib.pyplot as plt
import pandas as pd

# Datos proporcionados
data = {
    'Mechanism': [
        'Nucleic Acids Modification', 
        'Cyclic oligonucleotide-based antiphage signaling systems', 
        'Interference', 
        'Protein modification', 
        'Unknown', 
        'Targeting viral nucleic acids', 
        'Abortive',
        'NAD+ depletion'
    ],
    '%': [0.09, 0.09, 0.09, 0.09, 4.08, 33.84, 59.91,1.80]
}

# Crear DataFrame y ordenarlo en orden ascendente
df = pd.DataFrame(data)
df = df.sort_values(by='%', ascending=True)

# Crear el grfico de embudo
plt.figure(figsize=(10, 5))
bars = plt.barh(df['Mechanism'], df['%'], color='skyblue')

# Aadir etiquetas de porcentaje dentro de cada barra, ajustando para las barras grandes
for bar in bars:
    width = bar.get_width()
    text_position = width - 2 if width > 5 else width + 0.5  # Colocar el texto dentro si la barra es ancha

    plt.text(
        text_position,  # Posicin del texto
        bar.get_y() + bar.get_height() / 2,  # Centrar el texto verticalmente
        f'{width:.2f}%',  # Texto con el valor del porcentaje
        va='center',
        ha='center' if width > 5 else 'left',  # Alinear a la izquierda si est fuera, centrado si est dentro
        color='black',  # Color del texto
        fontsize=10
    )

# Configurar el ttulo y las etiquetas de los ejes
plt.xlabel('Percentage (%)')
plt.title('Distribution of defense mechanisms')
plt.tight_layout()

# Guardar la figura
plt.savefig('funnel_chart.png')

# Mostrar la figura
plt.show()
```
